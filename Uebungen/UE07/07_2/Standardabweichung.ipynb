```python
import numpy as np
import matplotlib.pyplot as plt

punkte4ahit = np.array([30, 31, 32, 32, 33])
punkte4bhit = np.array([28, 30, 32, 34, 34])

durchschnitt_4a = np.mean(punkte4ahit)
durchschnitt_4b = np.mean(punkte4bhit)

std_4a = np.std(punkte4ahit)
std_4b = np.std(punkte4bhit)

print(f"4AHIT - Mittelwert: {durchschnitt_4a:.2f} Punkte, Standardabweichung: {std_4a:.2f} Punkte")
print(f"4BHIT - Mittelwert: {durchschnitt_4b:.2f} Punkte, Standardabweichung: {std_4b:.2f} Punkte")
```

```python
plt.boxplot([punkte4ahit, punkte4bhit], vert=True, labels=['Punkte 4 AHIT', 'Punkte 4 BHIT'])

plt.title('Boxplot Punkte 4AHIT und 4BHITM')
plt.ylabel('Punkte')

plt.show()
# Whisker
```
